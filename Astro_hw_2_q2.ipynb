```python
import numpy as np
import plotly.graph_objs as go
import plotly.animation as animation

import numpy as np
import matplotlib.pyplot as plt

h=6.626*1e-34 #Planck constant in J/Hz
c=2.998*1e+8 #speed of light in m/s
k=1.381*1e-23 #Boltzmann constant in J/K
T=[4000,5000,6000] #Temperatures in K
e=2.72  #Euler's number
f=np.linspace(1e+1,2e+15,100000) #Frequency array in Hz

a=2*h/(c**2) #Constants for writing the function easily
b=h/k

#A function for Planck's Law
def intensity(freq,temp):
    I=(a*(freq**3))/((e**(b*freq/temp))-1)
    return I

I1=[]
I2=[]
I3=[]
for d in f:
    I1.append(intensity(d,T1))
    I2.append(intensity(d,T2)) 
    I3.append(intensity(d,T3))
    
#Max frequency values from Wien Displacement Law  
f_peak=[T1*5.879e10, T2*5.879e10, T3*5.879e10] 

#Intensity values for maximum frequencies
peak=[intensity(f_peak[0],T[0]), intensity(f_peak[1],T[1]),intensity(f_peak[2],T[2])]
```

```python
plt.plot(f,I1, color='deeppink', label='$T_1$=4 000 K')
plt.plot(f,I2, color='orange', label='$T_2$=5 000 K')
plt.plot(f,I3, 'g',label='$T_3$=6 000 K')
font1 = {'family':'serif','color':'darkblue','size':18}
font2 = {'family':'serif','color':'darkblue','size':15}
plt.xlabel("Frequency(Hz)", fontdict=font2)
plt.ylabel("Intensity(J/$m^2$)", fontdict=font2)
plt.title("Blackbody Radiation" ,fontdict = font1)
plt.plot(f_peak,peak, 'o-c',label="Wien's Line")
plt.legend(loc='upper right')
plt.text(f_peak[0],peak[0],"$f_1$",size=13)
plt.text(f_peak[1],peak[1],"$f_2$",size=13)
plt.text(f_peak[2],peak[2],"$f_3$",size=13)
plt.grid()
plt.show()
```
